**Configuration Variables**

```python
# Load the YAML file
with open('config.yaml', 'r') as file:
    config = yaml.safe_load(file)

# Function to set variables globally and store their names
def set_variables(config, prefix='', var_dict={}):
    for key, value in config.items():
        if isinstance(value, dict):
            set_variables(value, prefix + key + '_', var_dict)
        else:
            globals()[prefix + key] = value
            var_dict[prefix + key] = value
    return var_dict

# Set variables globally and get a dictionary of the set variables
set_vars = set_variables(config)

# Print all the variables that were set
print("Variables set from YAML file:")
for var_name, var_value in set_vars.items():
    print(f"{var_name}: {var_value}")
```

## <font color="#62b6cb"> 1. Data augmentation <a name="id1"></a>

### <font color="#62b6cb"> 1.1 Apply transformations <a name="id11"></a>

The data augmentation function requires that the annotation JSON to be inside the folder of the images. We copy the annotations inside the folders and later we will delete it when it is not needed anymore (only for training).

```python
shutil.copy(os.path.join(path_annotations,train_annotations_name), os.path.join(path_train_train,"annotations.json"))
```

```python
# Augmentation of train set

# input parameters clodsa
problem = "instance_segmentation"
annotation_mode = "coco"
input_path = path_train_train + '/' # you have to put the / at the end for opening the desired folder
generation_mode = "linear"

# output parameters clodsa
output_mode = "coco"
output_path = path_train_aug + '/'   # you have to put the \\ at the end for opening the desired folder


# creating output folder
if not os.path.exists(output_path):
    os.makedirs(output_path)

# create our augmentor object.
augmentor = createAugmentor(problem,annotation_mode,output_mode,generation_mode,input_path,{"outputPath":output_path})
# define a transformer generator.
transformer = transformerGenerator(problem)

# techniques applied to the photos

# rotation
for angle in [90,180]:
    rotate = createTechnique("rotate", {"angle" : angle})
    augmentor.addTransformer(transformer(rotate))

# flip
flip = createTechnique("flip",{"flip":1})
augmentor.addTransformer(transformer(flip))

# Gaussian noise
Gnoise = createTechnique("gaussian_noise", {"mean" : 0,"sigma":10})
augmentor.addTransformer(transformer(Gnoise))

augmentor.applyAugmentation()   
```

```python
# Apply identation  to the json of the augmented data
with open(os.path.join(output_path, 'annotation.json'), "r") as f:
    data = json.load(f)

# Guardar el archivo JSON con indentaci√≥n
with open(os.path.join(output_path, 'annotation.json'), "w") as f:
    json.dump(data, f, indent=4)
```
